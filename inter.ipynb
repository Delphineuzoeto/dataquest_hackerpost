```python

#Exploring Price by Brand 
brands_counts = autos["brand"].value_counts(normalize=True)
common_brands = brands_counts[brands_counts > .05].index
```

```python
brands_mean_prices = { }


for car_brands in common_brands:
    selected_brand = autos[autos["brand"] == car_brands]
    mean_price = selected_brand["price"].mean()
    brands_mean_prices[car_brands] = int(mean_price)
brands_mean_prices
```

```python
bmp_series = pd.Series(brands_mean_prices)
pd.DataFrame(bmp_series, columns=["mean_price"])
```

```python
# extract the brands name
autos["new_name"] = autos["name"].str.split("_").str[0]
autos["new_name"] = autos["new_name"].str.lower()


# Check the result
print(autos["new_name"].head(20))

```

```python
# Find the most common brand/model combinations

# for brands
autos["brandss"] = autos["name"].str.split("_").str[0].str.lower()
# for models
autos["models"] = autos["name"].str.split("_").str[1].str.lower()
# combine model and brand
autos["brands_model"] = autos["brandss"] + "_" + autos["models"]

autos["brands_model"].value_counts().head(10)

```

```python
# 1. CREATE MILEAGE GROUPS
bins = [0, 50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000, 500000]

labels = ["0-50k", "50k - 100k", "100k-150k", "150k-200k", "200k-250k", "250k-300k","300k-350k", "350k-400k", "400k-500k"]

# create a new column with mileage groups

autos["odometerkm_grps"] = pd.cut(autos["odometer_km"], bins=bins, labels=labels)


# 2. AGGREAGATE AVERAGE PRICE BY GROUP
#  Group by mileage group and calculate price by group

avg_price_by_mileage = autos.groupby("odometerkm_grps", observed=True)["price"].mean()

# sort for readability

avg_price_by_mileage = avg_price_by_mileage.sort_index()
print(avg_price_by_mileage)
```

```python
avg_price_by_damage = autos.groupby("unrepaired_damage")["price"].mean()
print(avg_price_by_damage)
```

```python
price_no_damage = avg_price_by_damage["No"]
price_with_damage = avg_price_by_damage["yes"]

difference = price_no_damage - price_with_damage
percentage_diff = (difference / price_no_damage) * 100

print(f"Damaged cars are on average {difference:.2f} cheaper ({percentage_diff:.1f}% less) than non-damaged cars.")
```
