Analyzing tables using object oriented python and dates objects.


```python
from csv import reader


# 1. Display the first five rows of hn
with open("hacker_news.csv") as file:
    hn = list(reader(file))
# hn = hn[:5]
# print(hn)

# 2.
headers = hn[0]
hn = hn[1:]

print(headers)
# print("\n", hn)

# 3.

ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    num_comments = [4]
    if title.startswith("Ask HN"):
        ask_posts.append(row)
    elif title.startswith("Show HN"):
        show_posts.append(row)
    else:
        other_posts.append(row)
        
# 4: total no of comments in ask posts

total_ask_comments = 0
for posts in ask_posts:
    try:
        
        num_comments = int(posts[4])
        total_ask_comments += num_comments
    except ValueError:
        continue
        
avg_ask_comments = round(total_ask_comments / len(ask_posts)) 
    
print("\n","total ask comment:",  total_ask_comments)
print("\n", "Average ask comment:", avg_ask_comments)
    
#  total no of comments in show post

total_show_comments =  0
for posts in show_posts:
    try:
        num_of_show_comments = int(posts[4])
        total_show_comments += num_of_show_comments
    except ValueError:
        continue
        
avg_show_comments = num_of_show_comments / len(show_posts)
print(avg_show_comments)
    
if avg_show_comments > avg_ask_comments:
    
    print("show posts receive more comments on average")
else:
    print("ask posts receive more comments on average")
 
  
    
# print(len(ask_posts), ask_posts)
# print(len(show_posts))
# print(len(other_posts))
```

```python
# 5 Finding the Number of Ask Posts and Comments by Hour Created
import datetime as dt

result_list =  [ ]

for row in ask_posts:
    created_at = row[6]
    # print(created_at)

   
    try:
        num_comments = int(row[4])
        date_obj = dt.datetime.strptime(created_at, "%m/%d/%Y %H:%M")
        hour = date_obj.strftime("%H:%M")
        result_list.append([hour, num_comments])
    except ValueError:
        continue


counts_by_hour = { }
comments_by_hour = { }

for row in result_list:
    hour = row[0]
    num_comments = row[1]
   
   
    if hour not in counts_by_hour:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = num_comments
    else:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += num_comments
        
    
print("Counts by hour:", counts_by_hour)
print("Comments by hour:", comments_by_hour)
```

```python

# Calculating the Average Number of Comments for Ask HN Posts by Hour
avg_by_hour = [ ]
for hour in comments_by_hour:
    avg = comments_by_hour[hour] / counts_by_hour[hour]
    avg_by_hour.append([hour, avg])
print(avg_by_hour)
    
    
  
```
